### 2-1. 상대습도

```python
RH_day_mean = train['RH'].groupby(train['Day']).mean()
RH_day_mean = pd.DataFrame(RH_day_mean).reset_index()
plt.figure(figsize=(15,4))
plt.title('Relative Humidity - Day Mean')
sns.lineplot(x='Day',y='RH',data = RH_day_mean,linewidth=1)
plt.show()
```

### 2-2. 온도

```python
T_day_mean = train['T'].groupby(train['Day']).mean()
T_day_mean = pd.DataFrame(T_day_mean).reset_index()
plt.figure(figsize=(15,4))
plt.title('Temperature - Day Mean')
sns.lineplot(x='Day',y='T',data=T_day_mean,linewidth=1)
plt.show()
```

### 2-3. 해가 떠있는 시간 (기준: GHI = DHI+DNI > 0)

```python
# 단순 Hour로만 비교
ghi = train.iloc[np.where(train['DHI'] + train['DNI'] > 0)[0],:] # GHI (= DHI + DNI)
ghi_min_hour = pd.DataFrame(ghi.groupby(['Day'])['Hour'].min()).reset_index()
ghi_max_hour = pd.DataFrame(ghi.groupby(['Day'])['Hour'].max()).reset_index()
ghi_minus_hour = ghi_max_hour.copy()
ghi_minus_hour['Hour'] = ghi_max_hour['Hour'] - ghi_min_hour['Hour']

plt.figure(figsize=(15,4))
plt.title('Sun')
sns.lineplot(x='Day',y='Hour',data=ghi_minus_hour,linewidth=1)
plt.show()
```

```python
# Hour & Minute 모두 비교
def time_minute(data):
    hour = data['Hour']
    if data['Minute'] == 0:
        minute = 0
    else: # ==30
        minute = 0.5
    return hour + minute

ghi = train.iloc[np.where(train['DHI'] + train['DNI'] > 0)[0],:].reset_index(drop=True) # GHI (= DHI + DNI)
ghi['time_float'] = [time_minute(ghi.iloc[i]) for i in range(len(ghi.index))]
ghi_min_hour = pd.DataFrame(ghi.groupby(['Day'])['time_float'].min()).reset_index()
ghi_max_hour = pd.DataFrame(ghi.groupby(['Day'])['time_float'].max()).reset_index()
ghi_minus_hour = ghi_max_hour.copy()
ghi_minus_hour['time_float'] = ghi_max_hour['time_float'] - ghi_min_hour['time_float']

plt.figure(figsize=(15,4))
plt.title('Sun')
sns.lineplot(x='Day',y='time_float',data=ghi_minus_hour,linewidth=1)
plt.show()
```

### 2-7. 이슬점 (Dew point)

```python
def dewpoint(data):
    rh = data['RH']
    t = data['T']
    gamma = math.log(rh/100) + (17.62*t)/(243.12*t)
    dew = (243.12*gamma)/(17.62-gamma)
    return dew

dew = train.copy()
dew['dew'] = [dewpoint(train.iloc[i]) for i in range(len(train.index))]
dewdew = pd.DataFrame(dew['dew'].groupby(dew['Day']).min()).reset_index()
plt.figure(figsize=(15,4))
plt.title('Dew - Day min')
sns.lineplot(x='Day',y='dew',data=dewdew,linewidth=1)
```

```python
# (산량일사량 - 직사일사량)을 비교
cloud = train.copy()
cloud['cloud'] = cloud['DHI'] - cloud['DNI']
sns.scatterplot(x='cloud',y='TARGET',data=cloud)
```

```python
# (산량일사량/직사일사량)을 비교
# 오른쪽의 증가하는 직선부분은 DNI=0인 부분 -> DNI가 0이기 때문에 DHI가 증가할수록 태양광 발정량도 증가했던 것
# 즉, 직선일사량(DNI)이 없으면 산량일사량(DHI)에 선형관계로 의존한다는 의미?
cloud = train.copy()
cloud['cloud'] = cloud['DHI']/(cloud['DNI']+0.001)
sns.scatterplot(x='cloud',y='TARGET',data=cloud)
```

```python
# 실제로 DNI=0이고 DHI!=0인 자료를 이용한 결과, DHI와 TARGET간 완벽한 선형관계 형성
e = train.iloc[np.where((train['DNI']==0)&(train['DHI']!=0))[0],:]
sns.scatterplot(x='DHI',y='TARGET',data=e)
```

```python
# GHI 그래프에서도 확인해본 결과 DHI > DNI인 경우, 태양광 발전량과 선형관계를 띰
# -> sns.scatterplot(x='GHI',y='TARGET',hue='more_DHI',data=e.iloc[np.where(e['more_DHI'] != 0)[0],:])을 통해서도 확인해봄
# DHI > DNI인 경우는 구름이 낀 흐린 날씨를 의미하는데, 
e = ghi2.copy()
e['more_DHI'] = [1 if e.DHI[i] > e.DNI[i] else 0 for i in range(len(e.index))]
sns.scatterplot(x='GHI',y='TARGET',hue='more_DHI',data=e)
```

```python
more_DNI = e.iloc[np.where(e['more_DHI'] == 0)[0],:].reset_index(drop=True)
T_dis = []
for i in range(len(more_DNI.index)):
    if more_DNI['T'][i] < -5:
        T_dis.append('~-5')
    elif (more_DNI['T'][i] >= -5) & (more_DNI['T'][i] < 5):
        T_dis.append('-5~5')
    elif (more_DNI['T'][i] >= 5) & (more_DNI['T'][i] < 15):
        T_dis.append('5~15')
    elif (more_DNI['T'][i] >= 15) & (more_DNI['T'][i] < 25):
        T_dis.append('15~25')
    else:
        T_dis.append('25~')
more_DNI['T_dis'] = T_dis
# sns.scatterplot(x='GHI',y='TARGET',hue='T_dis',data=more_DNI)
fig, axes = plt.subplots(2,3,figsize=(10,5))
sns.scatterplot(ax=axes[0,0],x='GHI',y='TARGET',data=more_DNI.iloc[np.where(more_DNI['T_dis'] == '~-5')[0],:])
sns.scatterplot(ax=axes[0,1],x='GHI',y='TARGET',data=more_DNI.iloc[np.where(more_DNI['T_dis'] == '-5~5')[0],:])
sns.scatterplot(ax=axes[0,2],x='GHI',y='TARGET',data=more_DNI.iloc[np.where(more_DNI['T_dis'] == '5~15')[0],:])
sns.scatterplot(ax=axes[1,0],x='GHI',y='TARGET',data=more_DNI.iloc[np.where(more_DNI['T_dis'] == '15~25')[0],:])
sns.scatterplot(ax=axes[1,1],x='GHI',y='TARGET',data=more_DNI.iloc[np.where(more_DNI['T_dis'] == '25~')[0],:])
```
